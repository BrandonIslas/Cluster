# Cluster No Jerarquico "Algoritmo K-means"

### Se utilizara el dataframe Spotify Dataset, el cual nos arroja parametros de las canciones que existen y las agrupa por genero musical, arrojando 15 generos distintos el dataset se obtuvo de: https://www.kaggle.com/adityaraut/spotify-dataset-predict-using-decisiontree/notebooks

### El objetivo de este estudio es analizar cuantos generos en base a las mismas categorias puede arrojar el algoritmo de cluster no jerarquico y comparar los resultados con los que ya cuenta el dataframe de Sopotify

Librerias utilizadas

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

Lectura de dataset

```python
dataframe= pd.read_csv('spotify.csv')
dataframe
```

Visualizacion de los generos con los que cuenta el dataset

```python
dataframe['genre'].unique()
```

Seleccion de las columnas objetivo

```python
X = dataframe.drop(['genre','song_name'], axis=1)
X.shape
```

```python
y= dataframe['genre']
y.shape
```

```python
X.describe()
```

Normalizacion de los valores para un mejor analisis

```python
data_normalizado= (X-X.min())/(X.max()-X.min())
data_normalizado.describe()
```

Libreria utilizada para el algoritmo de KMeans

```python
from sklearn.cluster import KMeans
```

A continuacion se realizo un estudio utilizando el analisis de Codo de Jambu, el cual nos permite seleccionar cuantos cluster deberian de ser los indicados para el analisis, ya que apartir de un numero de K dado la varianza no cambia y se pierde intension del analisis del algoritmo

```python
wcss=[]
for i in range(1,30):
    kmeans= KMeans(n_clusters=i, max_iter=300)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
plt.style.use('dark_background')
plt.figure(figsize=(15,5))
plt.plot(range(1,30), wcss, color='blue', marker='o', markerfacecolor='red', markersize=8)
plt.title("Codo de Jambu")
plt.xlabel("Numero de Clusters")
plt.ylabel("WCSS") #Indicador de que tan similares con los indivuos dentro de los clusters
plt.show()
```

En la grafica se puede observar que a partir de un K=10 (siendo estrictos K=5) el cambio en la tasa de error es muy pequeÃ±o por no decir nulo por lo que se tomo como punto de analisis la K=10

```python
modelo = KMeans(n_clusters=10)
```

Entrenamiento del modelo KMeans

```python
modelo.fit(X)
```

Obtencion de los centros arrojados por el modelo

```python
modelo.cluster_centers_
```

```python
modelo.labels_
```

Se agrego la comulna de los cluster a nuestro dataframe inicial

```python
dataframe['KMeans-Cluster']= modelo.labels_
dataframe
```

A continuacion se puede observar que solo basta de 10 generos para hacer la clasificacion de las canciones de spotifiy

```python
dataframe['KMeans-Cluster'].unique()
```
